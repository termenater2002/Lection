```python
# михин
class MixinSpeakable:
    def speak(self):
        print("я базарю")
        
class MixinAnimated:
    dance: bool = False
    def is_animated(self):
        print(f"я танцую и это {self.dance}")
        
class MixinFunny:
    def make_laugh(self):
        print("я иду на СВО")

class MixinActionable:
    def perform_action(self):
        print("казикоцианирую")

class MixinCollectible:
    is_collectible: bool = True
    def collect(self):
        print(f"меня собирают потные мужики и хранят у себя в подвале и это {self.is_collectible}")

class MixinPoseable:
    def pose(self):
        print("поз и р у ю")

class MixinCostumeWearable:
    costume: str = "базовый костюм каждого мужчины а именно трусы"
    def wear_costume(self):
        print(f"Ношу: {self.costume}")

# Базе
class BaseCharacter(MixinSpeakable, MixinAnimated, MixinFunny):
    name = "базовый чел"

class BaseHuman:
    name = "человек"

class BaseInActionCharacter(BaseCharacter, MixinActionable):
    name = "чел в действии"

class BaseFunkoPop(BaseCharacter, MixinCollectible):
    name = "фанкопоп"

class BaseCosplayer(BaseCharacter, BaseHuman, MixinPoseable, MixinCostumeWearable):
    name = "косплеер"

# уроды
class Shrek(BaseCharacter):
    name = "Шрек"

class PussInBoots(BaseCharacter):
    name = "Кот в сапогах"

class Donkey(BaseCharacter):
    name = "Осёл"

class JackHorner(BaseCharacter):
    name = "Джек Хорнер"

# Action
class ShrekInAction(BaseInActionCharacter):
    name = "Шрек в действии"

class PussInBootsInAction(BaseInActionCharacter):
    name = "Кот в сапогах в действии"

class DonkeyInAction(BaseInActionCharacter):
    name = "Осёл в действии"

class JackHornerInAction(BaseInActionCharacter):
    name = "Джек Хорнер в действии"

# Funko Pop
class ShrekFunkoPop(BaseFunkoPop):
    name = "Шрек Фанко Поп"

class PussInBootsFunkoPop(BaseFunkoPop):
    name = "Кот в сапогах Фанко Поп"

class DonkeyFunkoPop(BaseFunkoPop):
    name = "Осёл Фанко Поп"

class JackHornerFunkoPop(BaseFunkoPop):
    name = "Джек Хорнер Фанко Поп"

# Cosplayer
class ShrekCosplayer(BaseCosplayer):
    name = "Косплеер Шрека"

class PussInBootsCosplayer(BaseCosplayer):
    name = "Косплеер Кота в сапогах"

class DonkeyCosplayer(BaseCosplayer):
    name = "Косплеер Осла"

class JackHornerCosplayer(BaseCosplayer):
    name = "Косплеер Джека Хорнера"

# делаем
shrek = Shrek()
shrek.speak()
shrek.is_animated()
shrek.make_laugh()
print("_")
print("Action")

shrek_action = ShrekInAction()
shrek_action.speak()
print("_")
print("Funko Pop")

shrek_funko = ShrekFunkoPop()
shrek_funko.collect()
print("_")
print("Cosplayer")

shrek_cosplayer = ShrekCosplayer()
shrek_cosplayer.pose()
shrek_cosplayer.wear_costume()
```
